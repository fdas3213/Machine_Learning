```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torch.autograd import Variable
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**create dataset and dataloader**

```python
class MnistDataset(Dataset):
    """MNIST dataset"""
    def __init__(self, train_path, test_path = None):
        """
        Args:
            train_path(string): path to the csv file with features
            train_label_path(string): path to the csv file with train labels
        """
        self.data = np.array(pd.read_csv(train_path))
    
    def __len__(self):
        return len(self.data)
    
    def __getitem__(self, idx):
        X = self.data[idx][1:]
        y = self.data[idx][0]
        return {"feature": torch.from_numpy(X).float()/255, "label": np.array([y])}
```

```python
#take a look at first 3 data
mnist_dataset = MnistDataset(train_path="mnist_train.csv")
print(len(mnist_dataset))
for i in range(3):
    sample = mnist_dataset[i]
    print(i, sample['feature'].shape, sample['label'])
```

```python
batch_loader = DataLoader(mnist_dataset, batch_size = 128, shuffle=True)
sample = next(iter(batch_loader))
print(sample['feature'].shape, sample['label'].shape)
```

**A vanilla net**, reference from: https://course.fast.ai/lessons/lesson7.html

```python
class VanillaNet(nn.Module):
    def __init__(self, layers):
        super().__init__()
        self.layers = nn.ModuleList([nn.Linear(layers[i], layers[i+1]) for i in range(len(layers)-1)])
    
    def forward(self, x):
        for layer in self.layers:
            h = layer(x)
            x = F.relu(h)
        return F.log_softmax(h, dim=1)
```

**Setting training parameters and start training**

```python
EPOCHS = 3
criterion = nn.NLLLoss()
```

**1. Mini-batch SGD**

```python
#initialize network
VNet = VanillaNet([784, 100, 100, 10])
wbs = [(l.weight, l.bias) for l in VNet.layers]

#keep track of loss and setup learning rate
losses = list()
lr = 1e-3

#training 
for epoch in range(EPOCHS):
    print('EPOCH: {}'.format(epoch))
    for batch in iter(batch_loader):
        x, y = batch['feature'], batch['label']
        x, y = Variable(x), Variable(y)
        output = VNet(x)
        
        #compute loss
        loss = criterion(output, y.squeeze(-1))
        losses.append(loss)
        loss.backward()
        
        #weight update using SGD
        for w,b in wbs:
            w.data -= lr*w.grad.data
            b.data -= lr*b.grad.data
            w.grad.data.zero_()
            b.grad.data.zero_()
```

```python
sgd_losses = [np.log(l.item()) for l in losses]
fig, ax = plt.subplots(figsize = (14,8))
ax.plot(sgd_losses)
ax.set_title("mini-batch SGD")
```

**2. SGD with momentum**

```python
#initialize network
wbs = [(l.weight, l.bias) for l in VNet.layers]

#keep track of loss, set learning rate, and gamma to 0.9 as usual, and initialize velocity to 0
momentum_losses = list()
lr = 1e-3
gamma = 0.9
v_w = [0 for i in range(len(wbs))]
v_b = [0 for i in range(len(wbs))]

#training 
for epoch in range(EPOCHS):
    print('EPOCH: {}'.format(epoch))
    for batch in iter(batch_loader):
        x, y = batch['feature'], batch['label']
        x, y = Variable(x), Variable(y)
        output = VNet(x)
        
        #compute loss
        loss = criterion(output, y.squeeze(-1))
        momentum_losses.append(loss)
        loss.backward()
        
        #weight update using momentum
        for i, wb in enumerate(wbs):
            w, b = wb
            v_w[i] = gamma * v_w[i] + (1-gamma)* w.grad.data
            v_b[i] = gamma * v_b[i] + (1-gamma)* b.grad.data
            #perform updates
            w.data -= lr*v_w[i]
            b.data -= lr*v_b[i]
            #zero gradients
            w.grad.data.zero_()
            b.grad.data.zero_()
```

```python
sgd_momentum_losses = [np.log(l.item()) for l in momentum_losses]
fig, ax = plt.subplots(figsize = (14,8))
ax.plot(sgd_losses)
ax.plot(sgd_momentum_losses)
ax.legend(['sgd', 'sgd momentum'])
```

**3. Nesterov momentum**

```python
#initialize network 
wbs = [(l.weight, l.bias) for l in VNet.layers]

#keep track of loss and set up hyperparameters
nes_losses = list()
lr = 1e-3
gamma = 0.9

v_w = [0 for i in range(len(wbs))]
v_b = [0 for i in range(len(wbs))]

#training 
for epoch in range(EPOCHS):
    print('EPOCH: {}'.format(epoch))
    for batch in iter(batch_loader):
        x, y = batch['feature'], batch['label']
        x, y = Variable(x), Variable(y)
        output = VNet(x)
        
        #compute loss
        loss = criterion(output, y.squeeze(-1))
        nes_losses.append(loss)
        loss.backward()
        
        #weight update using momentum
        for i, wb in enumerate(wbs):
            w, b = wb
            w_original = w
            b_original = b
            #update ahead
            w.data += gamma * v_w[i]
            v_w[i] = gamma * v_w[i] - lr* w.grad.data
            
            b.data += gamma * v_b[i]
            v_b[i] = gamma * v_b[i] - lr* b.grad.data
            
            #perform updates
            w = w_original
            b = b_original
            w.data += v_w[i]
            b.data += v_b[i]
            #zero gradients
            w.grad.data.zero_()
            b.grad.data.zero_()
```

```python
nes_losses = [np.log(l.item()) for l in nes_losses]
fig, ax = plt.subplots(figsize = (14,8))
ax.plot(sgd_losses)
ax.plot(sgd_momentum_losses)
ax.plot(nes_losses)
ax.legend(['sgd', 'sgd momentum', 'nesterov momentum'])
```

**4. Adagrad**

```python
#initialize network
wbs = [(l.weight, l.bias) for l in VNet.layers]

#keep track of loss, set learning rate, and gamma to 0.9 as usual, and initialize velocity to 0
ada_losses = list()
lr = 1e-5
epsilon = 1e-6
grads_sum = [[torch.zeros(wb[0].size()), torch.zeros(wb[1].size())] for wb in wbs]

#training 
for epoch in range(EPOCHS):
    print('EPOCH: {}'.format(epoch))
    for batch in iter(batch_loader):
        x, y = batch['feature'], batch['label']
        x, y = Variable(x), Variable(y)
        output = VNet(x)
        
        #compute loss
        loss = criterion(output, y.squeeze(-1))
        ada_losses.append(loss)
        loss.backward()
        
        #weight update using adagrad
        #implementation follows from: http://cs231n.github.io/neural-networks-3/
        for i, wb in enumerate(wbs):
            w,b = wb
            #add square of gradients of weights 
            grads_sum[i][0] += np.square(w.grad.data)
            grads_sum[i][1] += np.square(b.grad.data)
            
            #perform updates 
            w.data -= lr*w.grad.data/(torch.sqrt(grads_sum[i][0]) + epsilon)
            b.data -= lr*b.grad.data/(torch.sqrt(grads_sum[i][1]) + epsilon)
            
            #zero gradients
            w.grad.data.zero_()
            b.grad.data.zero_()
```

```python
ada_losses = [np.log(l.item()) for l in ada_losses]
fig, ax = plt.subplots(figsize = (14,8))
ax.plot(sgd_losses)
ax.plot(sgd_momentum_losses)
ax.plot(nes_losses)
ax.plot(ada_losses)
ax.legend(['sgd', 'sgd momentum', 'nesterov momentum','adagrad'])
```

**5. adam**

```python
#initialize network
wbs = [(l.weight, l.bias) for l in VNet.layers]

#keep track of loss, and initialize some hyperparamters
adam_losses = list()
lr = 1e-3
eps = 1e-8
beta1 = 0.9
beta2 = 0.999
t = 0
m = [[torch.zeros(wb[0].size()), torch.zeros(wb[1].size())] for wb in wbs]
v = [[torch.zeros(wb[0].size()), torch.zeros(wb[1].size())] for wb in wbs]

#training 
for epoch in range(EPOCHS):
    print('EPOCH: {}'.format(epoch))
    for batch in iter(batch_loader):
        #keep track of time step
        t += 1
        
        x, y = batch['feature'], batch['label']
        x, y = Variable(x), Variable(y)
        output = VNet(x)
        
        #compute loss
        loss = criterion(output, y.squeeze(-1))
        adam_losses.append(loss)
        loss.backward()
        
        #weight update using adagrad
        #implementation follows from: http://cs231n.github.io/neural-networks-3/
        for i, wb in enumerate(wbs):
            w,b = wb
            #update m and v
            m[i][0] = beta1 * m[i][0] + (1-beta1) * w.grad.data
            m[i][1] = beta1 * m[i][1] + (1-beta1) * b.grad.data
            mt_w = m[i][0] / (1-beta1**t)
            mt_b = m[i][1] / (1-beta1**t)
            
            v[i][0] = beta2 * v[i][0] + (1-beta2) * w.grad.data * w.grad.data
            v[i][1] = beta2 * v[i][1] + (1-beta2) * b.grad.data * b.grad.data
            vt_w = v[i][0] / (1-beta2**t)
            vt_b = v[i][1] / (1-beta2**t)
            
            #perform updates 
            w.data -= lr*mt_w/(torch.sqrt(vt_w) + eps)
            b.data -= lr*mt_b/(torch.sqrt(vt_b) + eps)
            
            #zero gradients
            w.grad.data.zero_()
            b.grad.data.zero_()
```

```python
adam_losses = [np.log(l.item()) for l in ada_losses]
fig, ax = plt.subplots(figsize = (14,8))
ax.plot(sgd_losses)
ax.plot(sgd_momentum_losses)
ax.plot(nes_losses)
ax.plot(ada_losses)
ax.plot(adam_losses)
ax.legend(['sgd', 'sgd momentum', 'nesterov momentum', 'adagrad','adam'])
```
